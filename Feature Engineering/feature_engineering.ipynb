```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn.preprocessing
from sklearn.preprocessing import StandardScaler,MinMaxScaler
```

```python
file_name=f'sales_data2.csv'
if not os.path.exists(file_name):
    print(f'File: {file_name} Not Found')
    exit()
else:
    print(f'File: {file_name} Found')
```

```python
df=pd.read_csv('sales_data2.csv')
df.describe()
```

```python
print("shape of data (row,col):",df.shape)
```

```python
df.isnull().sum()
```

```python
df['Age']=df['Age'].fillna(df['Age'].mean())
df['Spending']=df['Spending'].fillna(df['Spending'].median())
df['Visits_Per_Month']=df['Visits_Per_Month'].mean()
df.isnull().sum()
```

```python
plt.bar(df['City'],df['Spending'])
plt.ylabel('spending')
plt.xlabel('city')
plt.show()
```

```python
corr=df.corr(numeric_only=True)
print(corr)
sns.heatmap(corr,annot=True)
plt.show()
```
